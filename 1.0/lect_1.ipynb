**Задача №1**

Дана строка в (кодировке UTF - 8)

Найти самый часто встречающийся в ней символ. Если несколько символов встречаются одинаково часто, то можно вывести любой.

***Решение №1***

Переберем все позиции и для каждой позиции в строке еще раз  переберем все позиции и в случае совпадения прибавим к счетчику единицу. Найдем максимальное значение счетчика.

***Сложность O(N^2)***

Т.к. два цикла, один вложенный.

N - длина строки

```python
s = input()
ans = ''
anscount = 0
for i in range(len(s)):
  nowcnt = 0
  for j in range(len(s)):
    if s[i] == s[j]:
      nowcnt += 1
  if nowcnt >= anscount:
    anscount = nowcnt
    ans = s[i]
print(ans, anscount)
```

***Решение №2***

Переберем все символы, встречающиеся в строке, а затем переберем все позиции и в случае совпадения прибавим к счетичку единицу. Найдем максимальное значение счетчика.

***Сложность O(Nk)***

N - длина строки

k -  кол-во различных символов

```python
s = input()
ans = ''
anscount = 0
for now in set(s):
  nowcnt = 0
  for j in range(len(s)):
    if now == s[j]:
      nowcnt += 1
  if nowcnt >= anscount:
    anscount = nowcnt
    ans = now
print(ans, anscount)
```

***Решение №3***

Заведем словарь, где ключом является символ, а значением - сколько раз он встретился. Если символ встретился впервые - создаем элемент словаря с ключом, совпадающим с этим символом и значение ноль. Прибавляем к элементу словаря с ключом, совпадающим с этим символом. единицу.

***Сложность O(N+k)***

N - длина строки

k -  кол-во различных символов

***Сложность O(N+k) = O(N)***

Т.к. k всегда меньше N -> O(2N) = O(N)

```python
s = input()
ans = ''
anscount = 0
dct = {}
for now in s:
  if now not in dct:
    dct[now] = 0
  dct[now] += 1
for key in dct:
  if dct[key] >= anscount:
    anscount = dct[key]
    ans = key
print(ans, anscount)
```

***Сумма последовательности***

```python
seq = list(map(int, input().split()))
seqsum = 0
for i in range(len(seq)):
  seqsum += seq[i]
print(seqsum)
```

***Максимум последовательности***

```python
seq = list(map(int, input().split()))
if len(seq) == 0:
  print('-inf')
else:
  seqmax = seq[0]
  for i in range(1, len(seq)):
    if seqmax < seq[i]:
      seqmax = seq[i]
  print(seqmax)
```

***Задача:***  даны три целых числа a, b, c.

Найдите все корни уравнения ax^2 + bx + c = 0 и выведете их в порядке возрастания.



```python
import math
a = -5
b = 4
c = 1
if a == 0:
  if b != 0:
    print(-c / b)
  if b == 0 and c == 0:
    print('Infinite number of solutions')
else:
  d = b ** 2 - 4 * a * c
  if d == 0:
    x1 = -b / (2 * a)
    print(x1)
  elif d > 0:
    x1 = (-b - math.sqrt(d)) / (2 * a)
    x2 = (-b + math.sqrt(d)) / (2 * a)
    if x1 < x2:
      print(x1, x2)
    else:
      print(x2, x1)
```
